## <span style="color:blue">OpenCV繪圖函式</span>
OpenCV至少有下面這些繪圖函式可以用來標明影像處理結果:
1. cv2.rectangle
2. cv2.circle
3. cv2.ellipse
4. cv2.line
5. cv2.arrowedLine
6. cv2.polylines
7. cv2.fillPoly
8. cv2.putText


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

img = cv2.imread('./test_data/highway.jpg')

if img is None:
    print('./test_data/highway.jpg not found')
    
font = cv2.FONT_HERSHEY_SIMPLEX

cx,cy = img.shape[1]//2,img.shape[0]//2

# draw rectangles
cv2.putText(img,'Rectangle',(0,cy), font, 0.4, (0,0,255),1)
# draw a blue rectangle
cv2.rectangle(img,(cx-10,cy-10),(cx+10,cy+10),(255,0,0),3)
# draw a filled green rectangle
cv2.rectangle(img,(cx-40,cy-10),(cx-20,cy+10),(0,255,0),-3)

# draw circles
cv2.putText(img,'circle',(0,cy-20), font, 0.4, (0,0,255),1)
cv2.circle(img,(cx-30,cy-20),10,(0,100,255),3)
cv2.circle(img,(cx,cy-20),10,(255,100,0),-1)

# draw lines
cv2.putText(img,'line',(0,cy-40), font, 0.4, (0,0,255),1)
cv2.line(img,(cx-40,cy-40),(cx,cy-40),(255,0,255),1)
cv2.arrowedLine(img,(cx+10,cy-40),(cx+50,cy-40),(255,0,255),1)

# draw polygons
cv2.putText(img,'polygon',(0,cy-60), font, 0.4, (0,0,255),1)
pts = np.array([[cx-50,cy-60],[cx-40,cy-55],[cx-10,cy-50],[cx-30,cy-60],[cx-40,cy-70]],dtype=int).reshape((-1,1,2))

cv2.polylines(img, [pts], True, (100,0,255))

pts[:,:,0] += 50
cv2.fillPoly(img, [pts], (100,0,255))

pts[:,:,0] += 50
cv2.polylines(img, [pts], False, (100,0,255))

# show images by plt.imshow or by cv2.imshow + cv2.waitKey
plt.figure(figsize = (img.shape[0]//30,img.shape[1]//30))
plt.axis('off')
plt.imshow(img[:,:,[2,1,0]])
plt.show()    
```
